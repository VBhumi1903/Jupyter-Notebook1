```python
import matplotlib.pyplot as pl
import numpy as np
import pandas as pd
```

```python
a=[1,2,3,4,5,6]
b=[15,17,18,19,20,21]
```

```python
a=np.array([11,12,13,14,15])
b=np.array([55,86,97,99,105])
pl.xlabel("Heigh")
pl.ylabel("age")
pl.title('Sample Data')
pl.plot(a,b,'r')
pl.show()
```

```python
a=np.array([11,12,13,14,15])
b=np.array([55,86,97,99,105])
pl.xlabel("Heigh")
pl.ylabel("age")
pl.title("Bar Graph")
pl.bar(a,b,color='black')
pl.show()
```

```python
a=np.array([22,25,19,17,24,6,2,3,33,58,31,60,45,78,37,56,44,89,85,77,55,56,46,48,47,45])
pl.hist(a,bins=[0,20,40,60,80,100],color="green")
pl.title("Histogram")
pl.show()
```
